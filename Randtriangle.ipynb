```julia
function b(σ=0.1,display=false) #creates a random triangle matrix
    A = fill(0,1000,1000);
    r() = σ * randn()

    w1, b1 = [1+r(), 0+r()], -333 + r() 
    w2, b2 = [0+r(), 1+r()], -333 + r()
    w3, b3 = [-1+r(), -1+r()], 1200 + r()
    #create the 3 weights and biases
    
    for i=1:1:1000, j=1:1:1000
        if((w1⋅[i,j]+b1>0) & (w2⋅[i,j]+b2>0) & (w3⋅[i,j]+b3>0))
            A[i,j]=1 #this line that needs to change.
        end
    end
    if(display==true)
        show(up.spy(sparse(A)))
    end
    return svdvals(A)[1:1:10]
end

```

```julia
function a(σ=0.1)
    A = fill(0,200,200);
    r() = 0.1 * randn()

    w1, b1 = [1+r(), 0+r()], -85 + r() 
    w2, b2 = [0+r(), 1+r()], -85 + r()
    w3, b3 = [-1+r(), -1+r()], 200 + r()
    #create the weights and baises
    
    for i=1:1:200, j=1:1:200
        if((w1⋅[i,j]+b1>0) & (w2⋅[i,j]+b2>0) & (w3⋅[i,j]+b3>0))
            A[i,j]=1 #this line that needs to change.
        end
    end
    
    return A,w1,b1,w2,b2,w3,b3 #returns weight + bias cause those needed for area caluclation
end

function area(w1,b1,w2,b2,w3,b3)
    W =[w1[1] w1[2] b1
        w2[1] w2[2] b2
        w3[1] w3[2] b3]

    C = inv(W')*det(W)
    #calculates the cofactors
    #calculate the area
    return det(W)^2/(C[1,3]*C[2,3]*C[3,3]*2)
end
```

```julia
A,w1,b1,w2,b2,w3,b3 = a()
```

```julia
area(w1,b1,w2,b2,w3,b3)
```

```julia
x=[]
y1=[]
y2=[]
y3=[]

for n=1:100
    A,w1,b1,w2,b2,w3,b3 = a()
    
    push!(x,area(w1,b1,w2,b2,w3,b3))
    push!(y1,svdvals(A)[1])
    push!(y2,svdvals(A)[2])
    push!(y3,svdvals(A)[3])

end

scatter(x,y1,xlabel="area",label="sigma1", title="200 by 200 matrices")
scatter!(x,y2,label="sigma2")
scatter!(x,y3,label="sigma3")

savefig("area vs singular values")
```

```julia
x=[]
y1=[]
y2=[]
y3=[]

for n=1:100
    A,w1,b1,w2,b2,w3,b3 = a()
    
    push!(x,area(w1,b1,w2,b2,w3,b3))
    push!(y1,svdvals(A)[1])
    push!(y2,svdvals(A)[2])
    push!(y3,svdvals(A)[3])

end

scatter(x.^(1/2),y1,xlabel="sqr(area)",label="sigma1",title="200 * 200 matrices")
scatter!(x.^(1/2),y2,label="sigma2")
scatter!(x.^(1/2),y3,label="sigma3")

savefig("sqr(area) vs singular values")
#mark this down as a odddity on my paper. An intresting oddity.
```
