```julia
using LinearAlgebra, Interact
```

```julia
A = fill(0,100,100);
```

```julia
i,j = 3,3; #starting location

n=5 #size of square. Length of sides

A[i.+(0:(n-1)),j.+(0:(n-1))].=1
```

```julia
svdvals(A)
```

```julia
using Pkg
Pkg.add("Interact")
```

```julia
@manipulate for n = 1:1:20, i = 5:5:30, j = 20:5:40
    A = fill(0,100,100);
    A[i.+(0:(n-1)),j.+(0:(n-1))].=1
    svdvals(A)[1:3]
end
```

```julia
@manipulate for n = 1:1:20 end
```
