```julia
n=33

csc.((1:2:(2n-1))*π/2/(2n+1))/2
csc((1)*π/2/(2n+1)) > csc((2n-1)*π/2/(2n+1))
y2=[]
y3=[]
for n=1:1:100
    push!(y2,csc((1)*π/2/(2n+1))/2)
    push!(y3,(2n+1)/(π))
end
Plots.scatter(x,y)
Plots.scatter!(x,y2)
Plots.scatter!(x,y3)
```
