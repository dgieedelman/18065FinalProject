```julia
using LinearAlgebra, Plots, UnicodePlots, SparseArrays
```

```julia
function b(σ=0.1,display=false) #creates a random triangle matrix
    A = fill(0,10,20);
    r() = σ * randn()

    w1, b1 = [1, 1+r()], -12
    w2, b2 = [1, -1+r()], 8 
    w3, b3 = [-1, 0], 8
    #create the 3 weights and biases
    
    for i=1:1:10, j=1:1:20
        if((w1⋅[i,j]+b1>0) & (w2⋅[i,j]+b2>0) & (w3⋅[i,j]+b3>0))
            A[i,j]=1 #this line that needs to change.
        end
    end
    if(display==true)
        show(UnicodePlots.spy(sparse(A)))
    end
    return A
end
L=b(0.001, true)
B=round.(pinv(L'*L))
println("")
for a = 1:20
    print(B[:,a])
    println(" ")
end
```
